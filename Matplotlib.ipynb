```python
# 3D lineplot 

x = [0,1,5,25]
y = [0,10,13,0]
z = [0,13,20,9]

plt.figure()
ax = plt.subplot(projection = '3d')
ax.scatter3D(x,y,z)
ax.plot3D(x,y,z)
```

```python
# 3D surface plot 
# example graph of x**2 + y**2

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize =(5,5))
ax = plt.subplot(projection = '3d')
a = ax.plot_surface(xx,yy,z,cmap = 'viridis')
fig.colorbar(a)
```

```python
# Contour plot
fig = plt.figure(figsize =(12,8))
ax = plt.subplot()
a = ax.contour(xx,yy,z,cmap = 'viridis')
fig.colorbar(a)
```

```python
# Filled contour
# Contour plot
fig = plt.figure(figsize =(12,8))
ax = plt.subplot()
a = ax.contourf(xx,yy,z,cmap = 'viridis')
fig.colorbar(a)
```
