```python
# Importing pandas library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing dataset
census = pd.read_csv('preliminary-data-of-national-population-and-housing-census-2021-english.csv')
```

```python
#preview of dataset- view first 5 rows
census.head()
```

```python
# Shape of the dataframe ,no of rows and columns present
census.shape
```

```python
# basic information about data types,storage and size
census.info()
```

```python
# Summary Statistics of dataset
census.describe()
```

```python
#Check for missing values
census.isnull().sum()
```

```python
census['District'].unique().tolist()
```

```python
# Local levels in each districts 
census['District'].value_counts().reset_index()
```

```python
# Find out total population,total male population and total female population of country
print('Total population : ',census['Total population'].sum())
print('Male population : ',census['Total Male'].sum())
print('Female population : ',census['Total Female'].sum())
```

```python
# Group by District and sum the Population
district_population = census.groupby('District')['Total population'].sum().reset_index().sort_values('Total population',ascending = False)
district_population
```

```python
# Find top 5 most & least populated districts
print('Top 5 most populated districts: \n\n ',district_population.head())
print('\n\n')
print('Top 5 least populated districts: \n\n ',district_population.tail())
```

```python
# Group by Local Level Name and sum the Population
local_population = census.groupby('Local Level Name')['Total population'].sum().reset_index().sort_values('Total population',ascending = False)
print(local_population.head())
print(local_population.tail())
```

```python
# Find out male female population ratio of country
census['Total Male'].sum()/census['Total Female'].sum()
```

```python
#Total number of household by districts 
household__by_district = census.groupby('District')['Total household number'].sum().sort_values()
print(household__by_district)
```

```python
# Average number of people per household
people_per_household = census['Total population'].sum()/census['Total household number'].sum()
people_per_household
```

```python
# Find districts where Total Male is greater than Total Female

# Group by District and sum the male and female populations
male_female_population = census.groupby('District')[['Total Male', 'Total Female']].sum().reset_index()

result = male_female_population[male_female_population['Total Male'] > male_female_population['Total Female']]

print(result)
```

```python
# Correlation between household,total population,male and female population
census[['Total household number', 'Total population','Total Male','Total Female']].corr()
```

```python
# Visualizing gender ratio by districts
# Calculate gender ratio
census['Gender_Ratio'] = (census['Total Male'] / census['Total Female']) * 100

# Sort by Gender Ratio
gender_ratio_sorted = census.sort_values(by='Gender_Ratio', ascending=False)

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(gender_ratio_sorted['District'], gender_ratio_sorted['Gender_Ratio'], color='blue')
plt.title('Gender Ratio (Males per 100 Females) by District')
plt.xlabel('District')
plt.ylabel('Gender Ratio')
plt.xticks(rotation=85)
plt.tight_layout()
plt.show()
```
