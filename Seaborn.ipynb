```python
import seaborn as sns
import plotly.express as px 
import matplotlib.pyplot as plt
```

```python
# Seaborn utility functions 
# 1) To view available dataset
sns.get_dataset_names()
```

```python
# 2) Load dataset 
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
titanic = sns.load_dataset('titanic')
flights = sns.load_dataset('flights')
```

### Relational plots
- scatter plot
- line plot
- figure level function = relplot()

```python
# Scatter plot 
# using axes level function 
sns.scatterplot(data = tips, x = 'total_bill',y = 'tip',hue = 'sex',size = 'size')
```

```python
# using figure level function - relplot()
sns.relplot(data = tips, x = 'total_bill',y = 'tip',hue = 'sex',size = 'size',kind = 'scatter')
```

```python
# Loading data from plotly 
country_record = px.data.gapminder()
country_record.sample()
```

```python
# line plot - most useful for time series data
# using axes level function 
#sns.lineplot(data = country_record, x = 'year',y = 'lifeExp',hue = 'continent',style = 'continent',size = 'continent').set_title('Life Expectancy Trend Continent(1950-2000)')
sns.lineplot(data = country_record, x = 'year',y = 'lifeExp',hue = 'continent').set_title('Life Expectancy Trend by  Continent(1950-2000)')
plt.show()
```

### Distribution plots 
- Histplot
- Kdeplot
- Rugplot
- figure level function = displot()

```python
# Histplot - univariate analysis
sns.histplot(data = tips, x='tip',hue = 'sex',element = 'step')
```

```python
# kdeplot - based on kernel density function
sns.kdeplot(data = tips, x = 'total_bill',fill = True,hue = 'sex')
plt.show()
```

```python
# rugplot - used as complement to other graph, generally no significance alone
sns.kdeplot(data = tips, x = 'total_bill')
sns.rugplot(data = tips, x = 'total_bill')
```

```python
# Using figure level function - displot()
sns.displot(data = tips, x='tip',hue = 'sex',element = 'step',kind = 'hist')
```

```python
sns.displot(data = tips, x='tip',hue = 'sex',kind = 'kde')
```

```python
# Bivariate histogram 
sns.displot(kind = 'hist',data = tips, x = 'total_bill',y = 'tip')
```

```python
# Bivariate kdeplot 
sns.displot(kind = 'kde',data = tips, x = 'total_bill',y = 'tip')
```

### Matrix plot
- Heatmap
- Clustermap
- no figure level functions

```python
# Heatmap - to plot grid based (matrix like) data 

# Creating grid structure data
temp = country_record[country_record['continent']=='Europe']
pivot = temp.pivot(index = 'country',columns = 'year',values='lifeExp')
temp
```

```python
plt.figure(figsize = (15,8))
sns.heatmap(pivot,annot = True,linewidth = 0.5,cmap ='viridis')
plt.title('Life Expectancy per year of European Countries')
```

```python
# Clustermap - heatmap + clustering 
sns.clustermap(pivot)
```

### Regression Plot 
- regplot
- lmplot
- residue plot

```python
# regplot - plots scatter plot and fits a regression line
sns.regplot(data = tips, x = 'total_bill',y = 'tip',marker = '*',color = 'black' ) # no hue parameter
plt.title('Bill vs Tip')
```

```python
# lmplot - hue present
sns.lmplot(data = tips, x = 'total_bill',y = 'tip',hue = 'sex' )
plt.title('Bill vs Tip per Gender')
```

```python
# resid plot - plots the error of regression fit
sns.residplot(data = tips, x = 'total_bill', y = 'tip')
```

### Multi Grid Plots
- FacetGrid
- PairGrid (vs pairplot)
- JointGrid(vs Jointplot)- bivariate and multivariate analysis at once 
  

```python
# Facetgrid
g = sns.FacetGrid(data = tips, col = 'day',row = 'time',hue = 'smoker')
g.map(sns.violinplot,'sex','total_bill')
g.add_legend()
```

```python
# pairplot
sns.pairplot(iris,hue = 'species')
```

```python
# PairGrid - adds more flexibility than pairplot 
g = sns.PairGrid(data = iris,hue = 'species')
g.map_diag(sns.histplot) #plots histogram along diagonal axes
g.map_upper(sns.kdeplot) #plots kde above than diagonal axes
g.map_lower(sns.scatterplot) #plots scatterplot below than diagonal axes
g.add_legend()
```

```python
# jointplot - univariate and bivariate visualization at once
sns.jointplot(data = tips, x = 'total_bill', y = 'tip',hue = 'sex')
```

```python
# using JointGrid
g = sns.JointGrid(data = tips, x = 'total_bill', y = 'tip',hue = 'sex')
g.plot(sns.scatterplot,sns.kdeplot)
```
